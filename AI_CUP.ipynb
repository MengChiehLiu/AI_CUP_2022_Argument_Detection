## **Dataset**

```python
train_raw = pd.read_csv('/content/drive/MyDrive/AI CUP/Batch_answers - train_data (no-blank).csv').iloc[:,:6]
# test_raw = pd.read_csv('/content/drive/MyDrive/AI CUP/')
```

```python
import string
train_raw['q_count'] = train_raw['q'].apply(lambda x: len(nltk.word_tokenize(''.join([i for i in x if i not in string.punctuation]))))
train_raw['r_count'] = train_raw['r'].apply(lambda x: len(nltk.word_tokenize(''.join([i for i in x if i not in string.punctuation]))))
train_raw['q_sum_count'] = train_raw['q\''].apply(lambda x: len(nltk.word_tokenize(''.join([i for i in x if i not in string.punctuation]))))
train_raw['r_sum_count'] = train_raw['r\''].apply(lambda x: len(nltk.word_tokenize(''.join([i for i in x if i not in string.punctuation]))))
train_raw['qr_sum_count'] = train_raw['q_sum_count'] + train_raw['r_sum_count']
```

### **TextRank**

```python
from gensim.summarization.summarizer import summarize
from gensim.summarization import keywords

def summarize_text(text):
    try:
        # summarized_text = summarize(text,ratio=0.4)
        summarized_text = summarize(text,ratio=0.2)
        if summarized_text!="":
            if summarized_text!=text:
                return summarized_text
            else:
                return text # return original text if summarized_text==text
        else:
            return text # return original text if summarized_text is empty
    except Exception as e:
        # print("Exception In summary-->")
        # print(str(e))
        return text # return original text if only has one sentence
```

```python
from rouge import Rouge
```

```python
'''
Calculate ROUGE score.
:parameter    
    :param predict: predicted summary (string or list) 
    :param summary: answer (string or list)
'''
def evaluate_summary(predict, summary):    
   rouge_score = Rouge()    
   scores = rouge_score.get_scores(predict, summary, avg=True)       
   score_1 = round(scores['rouge-1']['f'], 2)    
   score_2 = round(scores['rouge-2']['f'], 2)    
   score_L = round(scores['rouge-l']['f'], 2)    
   print("rouge1:", score_1, "| rouge2:", score_2, "| rougeL:", score_L)
```
